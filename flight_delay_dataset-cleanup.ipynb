# Preparing Flight Report dataset for merging (df)

The Monthly report data ('ONTIME_REPORTING_2020_01.csv') saved as df, has more Airport IDs('ORIGIN_AIRPORT_ID' and 'DEST_AIRPORT_ID')
350 unique values. Whereas the weather dataset has 97 unique ORIGIN_AIRPORT_IDs.

From merged weather data only the data of january need to be extracted First.

Both the dates and AIRPORT ID have to be considered while merging.


### Converting the month,day,year to date-time and reordering.

```python
df['YEAR'] = pd.Series([2019 for x in range(len(df.index))]) # Adding year column to dataset
df['YEAR']=df['YEAR'].astype(str)
df['DATE']=df['YEAR'].astype(str)+'/'+df['MONTH'].astype(str)+'/'+df['DAY_OF_MONTH'].astype(str) # combinig month day and year
```

```python
df['DATE']=pd.to_datetime(df['DATE']) # converting to date time
df.drop(['MONTH','DAY_OF_MONTH','YEAR'],axis=1,inplace=True) # dropping parts of date columns
```

```python
temp_cols=df.columns.tolist()    # Rearranging the columns
new_cols=temp_cols[30:] + temp_cols[0:30]
df=df[new_cols]
df.head(2)
```

```python
df=df.sort_values(by='DATE') # Arranging in ascending order.
df.head()
```

```python
df.reset_index(inplace=True,drop=True) # RESET THE INDEX
df.head(3)
```

```python
df.to_csv("cleaned_time_report_01.csv")
```
